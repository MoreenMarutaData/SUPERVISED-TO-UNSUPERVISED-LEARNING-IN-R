**mode**

````r
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
````

````r
# function to calculate the mode for the different columns
number <- c(colnames(df))
for (i in number ){
  ad_mode <- getmode(df[, i])
  print("The mode for")
  print(colnames(df[i]))
  print("is")
  print(ad_mode)
}
````

The modal values correspond to the tables we created right above.

````r
install.packages("caret")
library(caret)
dmy = dummyVars(" ~ .", data = df1)
ec.encod = data.frame(predict(dmy, newdata = ec.new))
str(ec.encod)
dim(ec.encod)
````

normalize

````r

normalize <- function(x){
  return ((x-min(x)) / (max(x)-min(x)))
}
normalize(ec.encod)
```
````

 Applying the K-means clustering algorithm with no. of centroids(k)=3

````r

result<- kmeans(ec.encod,10) 


result$size
result$cluster
result$centers
````

````r

par(mfrow = c(2,2), mar = c(5,4,2,2))


plot(ec.encod[c(1,2,3,4,5)], col = result$cluster)
```
````
