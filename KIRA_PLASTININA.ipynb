**mode**

````r
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
````

````r
# function to calculate the mode for the different columns
number <- c(colnames(df))
for (i in number ){
  ad_mode <- getmode(df[, i])
  print("The mode for")
  print(colnames(df[i]))
  print("is")
  print(ad_mode)
}
````

The modal values correspond to the tables we created right above.

````r
set.seed(0)

# plot between 2-15
k.max<- 15
data<- as.matrix(scale(df1))
wss<- sapply(1:k.max, function(k){kmeans(data, k, nstart=50, iter.max=15)$tot.withinss})
wss 
plot(1:k.max, wss, type="b", pch=19, frame=FALSE, xlab="no.of clusters k", ylab="total within clusters")

# function to compute total within-cluster sum of squares
#fviz_nbclust(df1, kmeans, method = "wss", k.max = 24) + theme_minimal() + ggtitle("the Elbow Method")
````

 Applying the K-means clustering algorithm with no. of centroids(k)=3 as observed above

````r
result<- kmeans(df1,3) 

result$size
result$cluster
result$centers
````

verifying results of cluster

````r
par(mfrow = c(2,2), mar = c(5,4,2,2))
plot(df1[c(1,2,3,4,5)], col = result$cluster)
````

## CHALLENGING THE SOLUTION

````r
# Compute dissimilarity matrix with euclidean distances
dis <- dist(df1, method = "euclidean")

# Hierarchical clustering using Ward's method
res.hc <- hclust(dis, method = "ward.D2" )

# Cut tree into 5 groups
grp <- cutree(res.hc, k =5)

# Visualize
plot(res.hc, cex = 0.6) 

# plot tree
rect.hclust(res.hc, k = 6, border = 2:5)
 # add rectangle
````
